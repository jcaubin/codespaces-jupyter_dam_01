```python
import chardet

def detect_encoding(file_path):
    with open(file_path, 'rb') as file:
        detector = chardet.universaldetector.UniversalDetector()
        for line in file:
            detector.feed(line)
            if detector.done:
                break
        detector.close()
    return detector.result['encoding']
```

```python
for url in csv_list:
    try:
        df = pl.read_csv(url, separator=';' ,encoding='cp1251', infer_schema_length=10000)
        print(url, df.schema.dtypes())
    except Exception as e:
        print(url, 'Error', e)
```
